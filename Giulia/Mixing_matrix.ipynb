# Matrice di mixing 

```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict, Counter
import seaborn as sns

G = nx.read_graphml(r"C:\Users\giuli\OneDrive\Documenti\UNI\MAGISTRALE\AnalVisual\AVRC_FinnishTwittersphereProject\Data\climate_19.graphml")

groups    = nx.get_node_attributes(G, 'group')      # 'A' o 'B'
hierarchy = nx.get_node_attributes(G, 'hierarchy')  # 'A_CORE', 'B_PERIPHERY', ...

mixing = nx.attribute_mixing_matrix(G, 'group', mapping={'A': 0, 'B': 1})
mixing_df = pd.DataFrame(mixing, index=['A', 'B'], columns=['A', 'B'])


fig, axes = plt.subplots(figsize=(6, 4))
sns.heatmap(mixing_df,    annot=True, fmt='.4f', cmap='Blues', ax=axes)
plt.suptitle("Mixing Matrix — comunità A vs B — Clima 19", fontsize=13)
plt.tight_layout()
plt.show()
```
